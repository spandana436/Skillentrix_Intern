```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# Download stopwords once

nltk.download('stopwords')

# Create objects

ps = PorterStemmer()

all_stopwords = set(stopwords.words('english'))

# Preprocessing function

def preprocess_text(text):

    # Remove special characters

    text = re.sub(
        '[^a-zA-Z]',
        ' ',
        str(text)
    )

    # Lowercase

    text = text.lower()

    # Split words

    text = text.split()

    # Remove stopwords + stemming

    text = [
        ps.stem(word)
        for word in text
        if word not in all_stopwords
    ]

    # Join words

    text = ' '.join(text)

    return text

# Preprocess training data

train_df['clean_review'] = train_df['0'].apply(
    preprocess_text
)

# Preprocess testing data

test_df['clean_review'] = test_df['0'].apply(
    preprocess_text
)

print(train_df.head())
```

```python
# Training preprocessing

train_df['clean_review'] = train_df['0'].apply(
    preprocess_text
)

# Testing preprocessing

test_df['clean_review'] = test_df['0'].apply(
    preprocess_text
)
```

```python
#Encode Labels
y_train = train_df['1'].map({
    'positive': 1,
    'negative': 0
})

y_test = test_df['1'].map({
    'positive': 1,
    'negative': 0
})
```

```python
print(train_df['0'][0])

print(train_df['1'][0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000
)

X_train = tfidf.fit_transform(
    train_df['clean_review']
).toarray()

X_test = tfidf.transform(
    test_df['clean_review']
).toarray()
```

```python
# label encode
y_train = train_df['1']

y_test = test_df['1']
```

```python
#MODEL TRAINING
from sklearn.naive_bayes import MultinomialNB

# Create model

model = MultinomialNB()

# Train model

model.fit(
    X_train,
    y_train
)

print("Model Training Completed")
```

```python
# PREDICTION
y_pred = model.predict(X_test)

print(y_pred[:10])
```

```python
# EVALUATION
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report
)

print("Accuracy:",
      accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:")

print(confusion_matrix(
    y_test,
    y_pred
))

print("\nClassification Report:")

print(classification_report(
    y_test,
    y_pred
))
```

```python
# TESTING CUSTOM REVIEWS
sample_review = """
This movie was absolutely fantastic and wonderful
"""

# Preprocess

clean_review = preprocess_text(
    sample_review
)

# Vectorize

review_vector = tfidf.transform(
    [clean_review]
).toarray()

# Predict

prediction = model.predict(
    review_vector
)

if prediction[0] == 1:
    print("Positive Review")
else:
    print("Negative Review")
```

```python
# NEGATIVE REVIEW
sample_review = """
Worst movie ever. Completely boring.
"""

clean_review = preprocess_text(
    sample_review
)

review_vector = tfidf.transform(
    [clean_review]
).toarray()

prediction = model.predict(
    review_vector
)

if prediction[0] == 1:
    print("Positive Review")
else:
    print("Negative Review")
```
